```python
import pandas as pd
fl=pd.read_csv("/content/drive/MyDrive/Assignment-1_Data.csv",delimiter=';')
fl
```

```python
#Items bought more
max= fl.groupby("Itemname")["Quantity"].transform("sum")
fl.loc[max.idxmax()]["Itemname"]
```

```python
#customer who buys more
fl.loc[fl.groupby("CustomerID")["Quantity"].transform("sum").idxmax()][["CustomerID"]]
```

```python
#customer who buys for more price
#fl['Price'] = fl['Price'].str.replace(',', '').astype(float)
fl.loc[fl.groupby("CustomerID")["Price"].transform("sum").idxmax()]["CustomerID"]
fl.groupby("CustomerID")["Price"].sum().sort_values(ascending=False)
```

```python
#customer bought recently
fl=fl.dropna(subset=["CustomerID"])
fl.sort_values(by=["Date"],ascending=False).head(50)
```

```python
#customer buys frequently
fl.groupby("CustomerID")["BillNo"].nunique().sort_values(ascending=False)
```
